```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing titanic csv file

```python
titanic = pd.read_csv("train.csv")
titanic.head()
```

# Step 1 : Finding and filling the missing values

```python
titanic.isnull().sum()
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False);
```

```python
titanic.drop('Cabin',axis = 1,inplace=True)
titanic.head()
```

```python
titanic['Pclass'].unique()
```

```python
titanic.shape
```

```python
titanic.head(50)
```

```python
mean_age = titanic.groupby('Pclass').mean()['Age']
```

```python
mean_age
```

```python
#Ready to fill the missing values in Age:
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[0]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 39
        elif Pclass==2:
            return 30
        else:
            return 26
    else:
        return Age
```

```python
titanic['Age'] = titanic[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False);
```

```python
titanic.dropna(inplace=True)
```

```python
titanic['Fare']=titanic['Fare'].fillna(titanic['Fare'].mean())
```

```python
titanic.dropna(inplace = True)
```

```python
titanic.isnull().sum()
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False);
```

```python
titanic.dropna(inplace=True)
```

# Step 2 converting Category data to numerical data

```python
#converting Age and Embarked Columns categorical data into numerical data 
```

```python
Sex = pd.get_dummies(titanic['Sex'])
Sex.head()
```

```python
Sex = pd.get_dummies(titanic['Sex'],drop_first = True)
Sex.head()
```

```python
Embark = pd.get_dummies(titanic['Embarked'],drop_first =True)
Embark.head()
```

```python
titanic.head()
```

```python
titanic.drop(['Embarked','PassengerId','Ticket'],axis=1,inplace = True)
```

```python
titanic.head()
```

```python
titanic = pd.concat([titanic,Sex,Embark],axis=1)
titanic.head()
```
